```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv")
```

```python
df.head(5)
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df.columns = headers
```

```python
df.head(5)

```

```python
df.replace("?", np.NaN, inplace=True)
df.head(5)
```

```python
missig_data = df.notnull()
missig_data.head(10)
```

```python
for column in missig_data.columns.values.tolist():
    print(column)
    print(missig_data[column].value_counts())
    print("")
```

```python
avg_nom_loss = df["normalized-losses"].astype(float).mean(axis=0)
print("Avarage normalized losses: ", avg_nom_loss)
```

```python
df["normalized-losses"].replace(np.NaN,avg_nom_loss, inplace=True)
df.head(5)
```

```python

avg_bore = df["bore"].astype(float).mean(axis=0)
print("avarage strokes is : " , avg_bore)
```

```python
df["bore"].replace(np.NaN,avg_bore, inplace=True)
```

```python
avg_stroke = df["stroke"].astype(float).mean(axis=0)
print("Avarage Stroke ",avg_stroke)
```

```python
df["stroke"].replace(np.NaN,avg_stroke, inplace=True)
```

```python
avg_horsepower = df["horsepower"].astype(float).mean(axis=0)
print("Avarage horsepower ",avg_horsepower)
df["horsepower"].replace(np.NaN,avg_horsepower, inplace=True)
```

```python
avg_peak_rpm = df["peak-rpm"].astype(float).mean(axis=0)
print("Avarage peak-rpm ",avg_peak_rpm)
```

```python
df["peak-rpm"].replace(np.NaN,avg_peak_rpm, inplace=True)
```

```python
df["num-of-doors"].value_counts()
```

```python
df["num-of-doors"].value_counts().idxmax()
```

```python
df["num-of-doors"].replace(np.NaN,'four' , inplace=True)
```

```python
df.dropna(subset=['price'],axis=0,inplace=True)
df.reset_index(drop=True,inplace=True)
```

```python
df[["bore","stroke"]] = df[["bore","stroke"]].astype(float)
df[["normalized-losses"]] = df[["normalized-losses"]].astype(int)
df[["horsepower"]] = df[["horsepower"]].astype(float)
df[["peak-rpm"]] = df[["peak-rpm"]].astype(float)
```

```python
df["city = L/100"] = 235/df["city-mpg"]
df.head(5)
```

```python
df["highway-mpg"] = 235/df["highway-mpg"]
df.rename(columns={"highway-mpg":"highway-L/100"},inplace=True)
df.head()
```

```python
df['width'] = df['width']/df['width'].max()
df['length'] = df['length']/df['length'].max()
df['height'] = df['height']/df['height'].max()
df[["height","width","length"]].head()
```

```python
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head(5)
```
